```python
from sklearn.metrics import mean_squared_error as mse, mean_absolute_error as mae
```

ERROR CALCULATION : IT HELPS US TO COMPARE DIFFERENT MODELS

```python
#testing MSE
test_mse = mse(ytest, reg.predict(xtest)) #ytest is actual y, reg.predict(xtest) is actually predicted y
test_mse
```

```python
#testing MAE
test_mse = mae(ytest, reg.predict(xtest))
test_mse
```

```python
#trainin MSE
train_mse = mse(ytrain, reg.predict(xtrain))
train_mse
```

```python
#trainin MAE
train_mse = mae(ytrain, reg.predict(xtrain))
train_mse
```
